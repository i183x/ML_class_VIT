# Encoding

## Converting object to int/float

### Types of encoding:
1. One hot encoding
2. Label encoding
3. Manual Encoding

```python
df.info()
```

```python
# One hot encoding
pd.get_dummies(df['model'])
```

```python
d = {'name':['a', 'b', 'c', 'd', 'e'], 'Age':[19,22,18,19,59], 'Gender' : ['M', 'F', 'M', 'M', 'F']}
df3 = pd.DataFrame(d)
df3['Salary'] = [34000, 50000, 45000, 30000, 30000]
df3
```

```python
one_hot = pd.get_dummies(df3['Gender'])
one_hot
```

```python
df4 = df3.join(one_hot)
df4
```

```python
df_enc = pd.get_dummies(df['model'])
df_enc
```

```python
df_new = df.join(df_enc)
df_new
```
