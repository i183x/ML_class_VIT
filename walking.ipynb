### File path

```python
jan_path = '/content/drive/MyDrive/2022_walking_data/jan.pdf'
mar_path = '/content/drive/MyDrive/2022_walking_data/mar.pdf'
jun_path = '/content/drive/MyDrive/2022_walking_data/jun.pdf'
jul_path = '/content/drive/MyDrive/2022_walking_data/jul.pdf'
aug_path = '/content/drive/MyDrive/2022_walking_data/aug.pdf'
sep_path = '/content/drive/MyDrive/2022_walking_data/sep.pdf'
oct_path = '/content/drive/MyDrive/2022_walking_data/oct.pdf'
nov_path = '/content/drive/MyDrive/2022_walking_data/nov.pdf'
dec_path = '/content/drive/MyDrive/2022_walking_data/dec.pdf'
```

### Data extraction

```python
jan_table = tabula.read_pdf(jan_path, pages='all')
mar_table = tabula.read_pdf(mar_path, pages='all')
jun_table = tabula.read_pdf(jun_path, pages='all')
jul_table = tabula.read_pdf(jul_path, pages='all')
aug_table = tabula.read_pdf(aug_path, pages='all')
sep_table = tabula.read_pdf(sep_path, pages='all')
oct_table = tabula.read_pdf(oct_path, pages='all')
nov_table = tabula.read_pdf(nov_path, pages='all')
dec_table = tabula.read_pdf(dec_path, pages='all')
```

### DataFrame creation

```python
df_jan = jan_table[0]
df_mar = mar_table[0]
df_jun = jun_table[0]
df_jul = jul_table[0]
df_aug = aug_table[0]
df_sep = sep_table[0]
df_oct = oct_table[0]
df_nov = nov_table[0]
df_dec = dec_table[0]
```

## Data manipulation & working

```python
# Assign the first row as column names for each DataFrame
df_jan.columns = df_jan.loc[0]
df_mar.columns = df_mar.loc[0]
df_jun.columns = df_jun.loc[0]
df_jul.columns = df_jul.loc[0]
df_aug.columns = df_aug.loc[0]
df_sep.columns = df_sep.loc[0]
df_oct.columns = df_oct.loc[0]
df_nov.columns = df_nov.loc[0]
df_dec.columns = df_dec.loc[0]

# Drop the first row for each DataFrame
df_jan.drop(0, inplace=True)
df_mar.drop(0, inplace=True)
df_jun.drop(0, inplace=True)
df_jul.drop(0, inplace=True)
df_aug.drop(0, inplace=True)
df_sep.drop(0, inplace=True)
df_oct.drop(0, inplace=True)
df_nov.drop(0, inplace=True)
df_dec.drop(0, inplace=True)
```

### Removing all the NaN

```python
df_jan= df_jan.fillna(0)
df_mar= df_mar.fillna(0)
df_jun= df_jun.fillna(0)
df_jul= df_jul.fillna(0)
df_aug= df_aug.fillna(0)
df_sep= df_sep.fillna(0)
df_oct= df_oct.fillna(0)
df_nov= df_nov.fillna(0)
df_dec= df_dec.fillna(0)
```

### Converting Step count from object to integer

```python
# Convert 'Step count' column to integer for each DataFrame
df_jan['Step count'] = df_jan['Step count'].astype(int)
df_mar['Step count'] = df_mar['Step count'].astype(int)
df_jun['Step count'] = df_jun['Step count'].astype(int)
df_jul['Step count'] = df_jul['Step count'].astype(int)
df_aug['Step count'] = df_aug['Step count'].astype(int)
df_sep['Step count'] = df_sep['Step count'].astype(int)
df_oct['Step count'] = df_oct['Step count'].astype(int)
df_nov['Step count'] = df_nov['Step count'].astype(int)
df_dec['Step count'] = df_dec['Step count'].astype(int)
```

# Plotting graph

```python
dates = np.arange(1,31)
plt.figure(figsize=(10,5))

# Plot data for January
steps_jan = df_jan['Step count']
plt.plot(dates, steps_jan, 'o-g', label='Jan')

# Plot data for March
steps_mar = df_mar['Step count']
# plt.plot(dates, steps_mar, 'o-k', label='Mar')      #days were not exactly 30

# Plot data for June
steps_jun = df_jun['Step count']
plt.plot(dates, steps_jun, 'o-b', label='Jun')

# Plot data for July
steps_jul = df_jul['Step count']
plt.plot(dates, steps_jul, 'o-r', label='Jul')

# Plot data for August
steps_aug = df_aug['Step count']
plt.plot(dates, steps_aug, 'o-m', label='Aug')

# Plot data for September
steps_sep = df_sep['Step count']
plt.plot(dates, steps_sep, 'o-c', label='Sep')

# Plot data for October
steps_oct = df_oct['Step count']
plt.plot(dates, steps_oct, 'o-y', label='Oct')

# Plot data for November
steps_nov = df_nov['Step count']
# plt.plot(dates, steps_nov, 'o-p', label='Nov')      #days were not exactly 30

# Plot data for December
steps_dec = df_dec['Step count']
plt.plot(dates, steps_dec, 'o-', color='orange', label='Dec')


plt.title("Recorded Step Count of 2022")
plt.xlabel("Dates of the month")
plt.ylabel("Step Count")
plt.legend()
plt.show()
```
